```python
import pandas as pd

# Load data
df = pd.read_csv(r"C:\Users\ayush\OneDrive\Desktop\New folder\zodecode\stud\APPLICATIONS\Pic&Sign\Resume\Analyst\ecommerce-analytics-project\data\raw\ecommerce.csv", low_memory=False)

# Drop unnecessary columns
df.drop(columns=["Unnamed: 22", "index"], inplace=True, errors='ignore')

# Convert date column to datetime~~~~
df["Order Date"] = pd.to_datetime(df["Order Date"], errors='coerce')

# Handle missing values

# Fill categorical columns with 'Unknown'
cat_cols = ["Courier Status", "currency", "fulfilled-by", "promotion-ids"]
for col in cat_cols:
    df[col] = df[col].fillna("Unknown")

# Fill numeric column
df["Amount"] = df["Amount"].fillna(0)

# 4. Fix postal code (should not be float)
df["ship-postal-code"] = df["ship-postal-code"].fillna(0).astype(int)

# 5. Standardize text columns (remove spaces, lowercase)
text_cols = ["Status", "Fulfilment", "Sales Channel", "Category", "ship-city", "ship-state"]
for col in text_cols:
    # Clean column names
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print(df.columns)

# 6. Remove duplicates
df.drop_duplicates(inplace=True)

# 7. Handle invalid quantities (optional safety check)
df = df[df["qty"] > 0]

# 8. Reset index
df.reset_index(drop=True, inplace=True)

# 9. Save cleaned data
df.to_csv(r"C:\Users\ayush\OneDrive\Desktop\New folder\zodecode\stud\APPLICATIONS\Pic&Sign\Resume\Analyst\ecommerce-analytics-project\data\cleaned\ecommerce_cleaned.csv", index=False)

print("Data cleaned successfully!")
print(df.info())
```
